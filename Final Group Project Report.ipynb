### Dataset Description:
We will utilize the "Heart Disease" dataset, which includes patient demographic information, clinical attributes, and a binary target variable indicating the presence or absence of heart disease. 
The dataset is sourced from a reliable repository and has been widely used for cardiovascular research.
The dataset comprises 76 attributes, with most studies centered on a subset of 14 attributes. Primarily sourced from the Cleveland database, extensively utilized by machine learning (ML) researchers, it aims to predict the presence of heart disease. The "goal" field denotes heart disease presence, ranging from 0 (absence) to 4.  
(Janosi, Andras, Steinbrunn,William, Pfisterer,Matthias, and Detrano,Robert, 1988)

The relevant columns in this spreadsheet are: 
1. **age**: age
2. **sex**: sex (1 = male, 0 = female)
3. **cp**: chest pain type
4. **trestbps**: resting blood pressure in mmHg
5. **chol**: serum cholestoral in mg/dl
6. **fbs**: fasting blood sugar > 120 mg/dl? (1 = True, 0 = False)
7. **restecg**: resting electrocardiographic results
8. **thalach**: maximum heart rate achieved
9. **exang**: whether exercise induced angina (1 = True, 0 = False)
10. **oldpeak**: ST depression induced by exercise, relative to rest
11. **slope**: the slope of the peak exercise ST segment (1 = upslope, 2 = flat, 3 = downslope)
12. **ca**: number of major vessels (0-3) colored by flourosopy
13. **thal**: (3 = normal, 6 = fixed defect, 7 = reversable defect)
14. **num**: diagnosis of heart disease (1,2,3,4 = presence, 0 = no presence)

#### Setting up the environment : Loading the dataset and libraries

```r
# importing libraries
library(tidyverse)
library(tidymodels)
library(repr)
library(RColorBrewer)


# formatting 
options(repr.plot.width = 6, repr.plot.height = 6)
options(repr.matrix.max.rows = 6)
```

```r
#reading the data
url<-"https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"

heart_data <-read_csv(url, col_names=FALSE) |>
                rename(age=X1,
                       sex=X2,
                       cp=X3,
                       trestbps=X4,
                       chol=X5,
                       fbs=X6,
                       restecg=X7,
                       thalach=X8,
                       exang=X9,
                       oldpeak=X10,
                       slope=X11,
                       ca=X12,
                       thal=X13,
                       num=X14)

print("                                TABLE 1:  HEART DISEASE DATA SET                   ")
heart_data
```

#### Tidying the data : Cleaning and Wrangling the data
- In the dataset description, several variables including 'sex', 'cp', 'fbs', 'restecg', 'exang', 'thal', 'ca', and 'slope' are identified as categorical. Utilizing the mutate function in combination with as.factor(), these variables are converted into categorical types.
- 
Cleaning of the data involves converting missing values denoted as "?" to NA. Moreover, the variable "sex" containing values 0 and 1 is transformed into "male" and "female" respectively for enhanced readability
- 
For a clearer interpretation of the variable "num", its description indicates that values 1 through 4 denote the presence of heart disease, while 0 signifies absence. To facilitate analysis, a new variable is created using the ifelse() function to discern the presence or absence of heart disease based on the values of the "num" variable. 

```r
#categorical variables
heart_newdata <- heart_data |>
                 mutate(heart_disease_status = as.factor(ifelse(is.na(num), NA, (num > 0)))) |>
                    mutate(num = as_factor(num)) |>
                    mutate(sex = as.factor(sex), cp = as.factor(cp), 
                           fbs = as.factor(fbs), restecg = as.factor(restecg),
                           exang = as.factor(exang), thal = as.factor(thal),
                           ca = as.factor(ca), slope = as.factor(slope))
#missing values
heart_newdata[ heart_newdata == "?" ] <- NA

heart_newdata <- heart_newdata|> 
                mutate(heart_disease_status = ifelse(num == 0, 'ABSENT', 'PRESENT'))|>
                mutate(sex = ifelse(sex == 1, 'male', 'female')) |>
                mutate(heart_disease_status = as.factor(heart_disease_status))
               

print("                          TABLE 2 :    HEART DISEASE TIDY DATA                     ")
heart_newdata
```

#### Splitting our data into training and testing sets

Before we start building our model, we need to split our data into two parts: one for training and the other for testing We will use the training data to train our classifer and testing data to evaluate the performance of the model.  We want to predict a new column called`heart_disease_status`, so we'll split the data in a way that ensures there's a good mix of `heart_disease_status` values in both sets   .

We'll use a method called`initial_split` to do this. It divides our data so that 75% is used for training and 25% for testing. This split helps us train our model with lots of data while still having enough left over to check how well it works. In this case we have 228 rows for training and 75 rows for testing.Using `initial_split` also allows us to shuffle the data before splitting (removing bias and order) so that an equal proportion of each is in each sets. 

This is demonstrated in **TABLE 3** and **TABLE 4**.

```r
#splitting the dataframe into training and testing data
heart_split <- initial_split(heart_newdata, prop = 3/4, strata = num)

heart_training <- training(heart_split)
heart_testing <- testing(heart_split)

print("                                 TABLE 3:    TRAINING DATA                     ")
head(heart_training)
print("                                 TABLE 4:    TESTING DATA                     ")
head(heart_testing)
```

#### Summarizing the training data 
Now we can summarize the data, to see the average values of variables (non-factor) like chol (serum cholestoral in mg/dl), trestbps (resting blood pressure in mmHg) and thalach (maximum heart rate achieved) (predictor variables) and find the total number of patients. This is done by `group_by` and `summarize` to find the average of the cholesterol and maximum heart rate of patients.

```r
heart_summarized <- heart_training |>
                    group_by(heart_disease_status) |>
                    summarize(no_of_patients = n(), avg_chol = mean(chol),
                              avg_thalach = mean(thalach))

print("              TABLE 5:      SUMMARIZED DATA                      ")
heart_summarized 
```

**TABLE 5** :- 
From the data above, we can see that patients with higher cholesterol and lower maximum heart rate have heart disease compared to the people who don't have the disease.

#### Visualizing the realtionship between predictors
To visualize the relationship that cholesterol and maximum heart rate achieved have with a patient's diagnosis, we will use `ggplot` to generate a scatter plot of chol and thalach against eacho other with colouring for `heart_disease_status`. 

```r
#plotting the training data
heart_dataplot1 <- ggplot(heart_training, aes(x = thalach, y = chol, color = heart_disease_status)) +
                 geom_point() +
                 labs(x = "max heart rate reached", y = "cholesterol (mg/dl)",
                      color = "Heart Disease Status", title = "Serum cholesterol vs maximum heart rate achieved") +
                 scale_color_brewer(palette = "Set1") +
                 theme(text = element_text(size = 12))

print ("                        FIGURE 1                     ")
heart_dataplot1
```

**FIGURE 1** -
This chart indicates that individuals with heart disease (blue) typically exhibit lower maximum heart rates and higher cholesterol levels, while those without heart disease (red) tend to have higher maximum heart rates and a broader range of cholesterol levels.

Viewing the data in this manner enables us to not only confirm the findings of the previous summarized data (TABLE 5) but also observe the impact of outliers on the values it presented. Moreover, it provides insight into the dispersion of the different classes.

### Tuning the Model: Finding the best $k$ value

#### Creating the model
- **Create a Recipe** : create a recipe that uses `heart_training` to predict `heart_disease_status`, with `chol` and `thalach` as predictors.
- **Perform Cross-Validation** : Divide the dataset into 10 equally sized sections (folds), ensuring each fold maintains a proportional representation of `heart_disease_status`.
- **Workflow Establishment**: Establish a workflow integrating the initial two steps, substituting 'tune_grid' for 'fit'.
- **Iteration for Validation Sets**: Iterate through each validation set, considering values within the range of 1 to 100."

But first we can use `group_by` and `summarize` to find the percentage of malignant and benign classes in heart_training and we see about 54% of the training data are patients without heart disease and 45% are patients with heart disease, indicating that our class proportions are preserved when we split the data.

This can be seen in **TABLE 6** below.

```r
heartdisease_proportions <- heart_training |>
                      group_by(heart_disease_status) |>
                      summarize(n = n()) |>
                      mutate(percent = 100*n/nrow(heart_training))
print("      TABLE 6 : COMPARING PROPORTIONS        ")
heartdisease_proportions
```

Now, we can do the following steps to tune the model: 

```r
set.seed(1) #using set seed to produce reproducable results

#Creating the Recipe 
heart_recipe <- recipe(heart_disease_status ~ chol + thalach, data = heart_training) |>
   step_scale(all_predictors()) |>
   step_center(all_predictors())

#Model Specification
knn_tune <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) |>
            set_engine("kknn") |>
            set_mode("classification")


#Cross-Validation over 10 folds
heart_vfold <- vfold_cv(heart_training, v = 10, strata = heart_disease_status)

gridvals <- tibble(neighbors = c(1:100))

# Creating workflow
knn_results <- workflow() |>
      add_recipe(heart_recipe) |>
      add_model(knn_tune) |>
      tune_grid(resamples = heart_vfold, grid = gridvals) |>
      collect_metrics() |>  #Collecting the Metrics 
      filter(.metric == "accuracy") |>
      select(neighbors, mean) |>
      arrange(desc(mean))

print ("         TABLE 7: CHOOSING K  ")

knn_results
```

From the above table, it can be seen we get a $66%$ accuracy when we assign k as 63 or 64. We can graph a plot to get a better visualization. 

```r
cross_val_plot<- ggplot(knn_results, aes(x = neighbors, y = mean)) +
      geom_line() +
      labs(x = "Neighbors", y = "Accuracy") +
      scale_x_continuous(breaks = seq(0, 100, by = 5)) +  
      scale_y_continuous(limits = c(0.4, 1.0)) +        
      ggtitle("Choosing K")

print ( "                        FIGURE 2                           ")
cross_val_plot
```

We can see that choosing 63 or 64 does not make much of a difference. So we are choosing 71 as our number of neighbors, we have finished tuning the model.

### Building the Final Classification Model
Since we chose 63  as our k value, we can fit into our `heart_training` dataset, to build the model.

```r
set.seed(1) #using set seed to produce reproducable results

specification <- nearest_neighbor(weight_func="rectangular", neighbors = 63) |>
    set_engine("kknn") |>
    set_mode("classification")

fit <- workflow() |>
    add_recipe(heart_recipe) |>
    add_model(specification) |>
    fit(data = heart_training)

fit
```

#### Testing the Model

Now that we have built the model, let's test our performace of the classifier by using the testing portion of our data.

```r
set.seed(1) #using set seed to produce reproducable results

#predicting the model
predictions <- predict(fit, heart_testing) |>
               bind_cols(heart_testing)

#calculating the accuracy 
metrics<- predictions |>
    metrics(truth = heart_disease_status, estimate = .pred_class) |>
    filter(.metric == "accuracy") 
print("      TABLE 8 : ACCURACY        ")
metrics
```

We see that the classifier has accuracy of 73%. Let's take a a closer look at the confusion matrix to evaluate the results.

#### Results

Analyzing the results of the prediction model, We see that our classifer has accuracy of 72%. Let's now take a look at the confusion matrix using the 'conf_mat' function.

```r
#using a confusion matrix, to analyze the weak areas
conf_mat<- predictions |>
    conf_mat(truth = heart_disease_status, estimate = .pred_class)
print("      TABLE 9 :  CONFUSION MATRIX   ")
conf_mat
```

Now, looking at the confusion matrix, it shows that 22 were correctly diagnosed with heart disease while 35 were correctly diagnosed without heart disease. The weak areas have also been represented, specifically  15 were incorrectly diagonosed with heart disease and 6 were incorrectly diagnosed without heart disease. 

We can calculate the accuracy, precision and recall, 

```r
# Calculate accuracy
accuracy <- (35 + 22) / (35 + 15 + 6 + 22) * 100
accuracy

# Calculate precision
precision <- 22 / (22 + 6) * 100
precision

# Calculate recall (sensitivity)
recall <- 22 / (22 + 15) * 100
recall
```

Therefore, we have 73% of accuracy, 78% of precision and 59% of recall.

This is a pretty decent percentage, and our classifier does a pretty good job of predicting the heart disease status of a patient. Although medically speaking, it is better to reduce error possiblities to a maximum extent when it comes to predicting the diagnoses.
